```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import statsmodels.formula.api as sm
```

# Data Processing

## EU Data

```python
# import csv
eu_data = pd.read_csv('./data/eu_data.csv')

# drop unuse data 
eu_data.drop(columns=['country_code', 'source', 'new_cases', 'number_sequenced', 'percent_cases_sequenced', 'valid_denominator', 'number_sequenced_known_variant', 'percent_variant'], inplace=True)

#variant of interest
voi = ["B.1.1.7", "B.1.351", "P.1", "B.1.617.2"]
# voi = [alpha, beta, gamma, delta]

eu_data = eu_data[eu_data.variant.isin(voi) == True]
eu_data.dropna(subset = ['number_detections_variant'], inplace=True)
eu_data.reset_index(drop=True, inplace=True)
eu_data
```

```python
eu_data_alpha = eu_data[eu_data.variant == "B.1.1.7"]
eu_data_beta = eu_data[eu_data.variant == "B.1.351"]
eu_data_gamma = eu_data[eu_data.variant == "P.1"]
eu_data_delta = eu_data[eu_data.variant == "B.1.617.2"]
```

## World Data

```python
# Variant of interest
voi = ["Alpha", "Beta", "Gamma", "Delta"]

world_data = pd.read_csv('./data/world_data.csv')
world_data.drop(columns=['num_sequences_total', 'perc_sequences'], inplace=True)
world_data.dropna(subset=['num_sequences'], inplace=True)
world_data = world_data[world_data.variant.isin(voi) == True]
world_data.reset_index(drop=True, inplace=True)
world_data
```

```python
world_data_alpha = world_data[world_data.variant == "Alpha"]
world_data_beta = world_data[world_data.variant == "Beta"]
world_data_gamma = world_data[world_data.variant == "Gamma"]
world_data_delta = world_data[world_data.variant == "Delta"]
```

# Analysis

## EU Data

### Graph EU Cases

```python
# for EU data only
def data_clean(data, country):
    data_processed = data.copy()
    data_processed = data_processed[data_processed.country == country].sort_values(by = ['number_detections_variant'], ascending=False)
    data_processed = data_processed.drop_duplicates(subset = ["year_week"]).sort_index()
    data_processed = data_processed.set_index('year_week')
    return data_processed

def plot_graph(country, label):
    plt.plot(country.index, country.number_detections_variant, label=label)
```

```python
# obtain list of countries on EU data
country_eu = eu_data.copy()
country_eu = country_eu.drop(columns=['variant', 'number_detections_variant', 'year_week'])
country_eu = country_eu.drop_duplicates(subset = ['country'])
country_eu.reset_index(drop=True, inplace=True)
```

```python
# EU alpha variant cases
plt.figure(figsize=(20, 10))
plt.xticks(rotation=90)
plt.xlabel("Year-week")
plt.ylabel("Number of detected cases")
plt.title("Europe Alpha Variant Cases")
for i in country_eu['country']:
    plot_graph(data_clean(eu_data_alpha, i), i)
plt.legend()
```

```python
# EU beta variant cases
plt.figure(figsize=(20, 10))
plt.xticks(rotation=90)
plt.xlabel("Year-week")
plt.ylabel("Number of detected cases")
plt.title("Europe Beta Variant Cases")
for i in country_eu['country']:
    plot_graph(data_clean(eu_data_beta, i), i)
plt.legend()
```

```python
# EU gamma variant cases
plt.figure(figsize=(20, 10))
plt.xticks(rotation=90)
plt.xlabel("Year-week")
plt.ylabel("Number of detected cases")
plt.title("Europe Gamma Variant Cases")
for i in country_eu['country']:
    plot_graph(data_clean(eu_data_gamma, i), i)
plt.legend()
```

```python
# EU delta variant cases
plt.figure(figsize=(20, 10))
plt.xticks(rotation=90)
plt.xlabel("Year-week")
plt.ylabel("Number of detected cases")
plt.title("Europe Delta Variant Cases")
for i in country_eu['country']:
    plot_graph(data_clean(eu_data_delta, i), i)
plt.legend()
```

```python
# Overal graph for all EU variant of concern cases
plt.figure(figsize=(20, 10))
plt.xticks(rotation=90)
plt.xlabel("Year-week")
plt.ylabel("Number of detected cases")
plt.title("Europe COVID-19 Variant of Concern Cases")
plt.scatter(eu_data_alpha.year_week, eu_data_alpha.number_detections_variant, label="alpha")
plt.scatter(eu_data_delta.year_week, eu_data_delta.number_detections_variant, label="delta")
plt.scatter(eu_data_beta.year_week, eu_data_beta.number_detections_variant, label="beta")
plt.scatter(eu_data_gamma.year_week, eu_data_gamma.number_detections_variant, label="gamma")
plt.legend()
```

### Graph Observation Analysis

From the graph of each variant, it is possible to observe that the rise and drop in cases for a certain variant happens mostly at the same time frame throughout Europe. At the same time we can also observe that after the drop in cases, the variant is mostly no where to be found and replaced by another different variant.

While for the scattered graph, we can observe that the alpha COVID-19 mutation is the dominant strain in Europe before the number of cases drops and replaced by the delta variant.

___Thus, from all the graphs we can obtain that the pattern, location, and spreading of different variants is similar across Europe.___


### EU Regression and Correlation 

```python
# Remove duplicate data 
week = eu_data.copy()
week = week.drop(columns=['variant', 'number_detections_variant', 'country'])
week = week.drop_duplicates(subset = ['year_week'])
week.reset_index(drop=True, inplace=True)
```

```python
def ols_location_time(data):
    eu_data_var = data.copy()

    for i in range(len(week)):
        eu_data_var.replace(to_replace = week['year_week'][i], value = (i + 1), inplace= True)

    result = sm.ols(formula="number_detections_variant ~ country + year_week", data=eu_data_var).fit()
    print(result.summary())
    
    for i in range(len(country_eu['country'])):
        eu_data_var.replace(to_replace = str(country_eu['country'][i]), value = (i + 1), inplace= True)
    sns.heatmap(eu_data_var.corr(), vmin=-1, vmax=1, annot=True, cbar=False).set_title("Correlation Analysis", fontdict={'fontsize':12}, pad=12);
```

```python
# EU OLS Regression Result and correlation analysis
ols_location_time(eu_data)
```

```python
# EU alpha variant OLS Regression Result and correlation analysis
ols_location_time(eu_data_alpha)
```

```python
# EU beta variant OLS Regression Result and correlation analysis
ols_location_time(eu_data_beta)
```

```python
# EU gamma variant OLS Regression Result and correlation analysis
ols_location_time(eu_data_gamma)
```

```python
# EU delta variant OLS Regression Result and correlation analysis
ols_location_time(eu_data_delta)
```

#### Overal Analysis For the Whole EU data

**OLS Regression**

With most p<0.05 and p=0 in most cases, it has shown strong evidence that the location(country) and time(year_week) affect the spread of a certain variant and overall(all variant data combined). Some tests that result p>0.05 and have tvalue that close to 0, indicating there might be a lack of sample to have a similar result like the other.

**Correlation Analysis**

Based on this analysis it is further strengthen the fact that the time frame matter and shown on the eu_data_delta and eu_data heatmap. Therefore, we need to make further correlation analysis and separate the data based on a certain time frame.

```python
def corr_eu(data):
    eu_data_var = data.copy()

    for i in range(len(week)):
        eu_data_var.replace(to_replace = week['year_week'][i], value = (i + 1), inplace= True)
        
    for i in range(len(country_eu['country'])):
        eu_data_var.replace(to_replace = country_eu['country'][i], value = (i + 1), inplace= True)
    
    fig, axs = plt.subplots(nrows = 1, ncols = 3, figsize = (25,6))
    sns.heatmap(eu_data_var.loc[(eu_data_var['year_week'] >= 1) & (eu_data_var['year_week'] <= 18)].corr(), vmin=-1, vmax=1, annot=True, ax = axs[0], cbar=False).set_title("Week 1 to 18", fontdict={'fontsize':20}, pad=12);
    sns.heatmap(eu_data_var.loc[(eu_data_var['year_week'] >= 19) & (eu_data_var['year_week'] <= 36)].corr(), vmin=-1, vmax=1, annot=True, ax = axs[1], cbar=False).set_title("Week 19 to 36", fontdict={'fontsize':20}, pad=12);
    sns.heatmap(eu_data_var.loc[(eu_data_var['year_week'] >= 37) & (eu_data_var['year_week'] <= 54)].corr(), vmin=-1, vmax=1, annot=True, ax = axs[2], cbar=False).set_title("Week 37 to 54", fontdict={'fontsize':20}, pad=12);
```

```python
# EU correlation analysis during different time frame
corr_eu(eu_data)
```

```python
# EU ALPHA VARIANT correlation analysis during different time frame
corr_eu(eu_data_alpha)
```

```python
# EU BETA VARIANT correlation analysis during different time frame
corr_eu(eu_data_beta)
```

```python
# EU GAMMA VARIANT correlation analysis during different time frame
corr_eu(eu_data_gamma)
```

```python
# EU DELTA VARIANT correlation analysis during different time frame
corr_eu(eu_data_delta)
```

#### Overal Analysis For EU data Heatmap

After observing the heatmap, noticed that there is a pattern that applies to all COVID-19 variants of concern. 

Observed That the year_week to num_detections_variant correlation is higher when the variant is about to rise and after the outbreak. While the country to num_detections_variant correlation coefficient is highest during the outbreak.

Indicating a relation between time with number of cases before and after the outbreak. While the location(country) is related to the number of cases during the outbreak. 

Further strengthening the fact that **the outbreak of each variant mostly happened at different locations during the same period of time with the same pattern of spreading.**

### World Data Regression and Correlation 

```python
# drop duplicate data
week = world_data.copy()
week = week.drop(columns=['variant', 'num_sequences', 'location'])
week = week.drop_duplicates(subset = ['date'])
week.reset_index(drop=True, inplace=True)
```

```python
# get list of countries 
location_world = world_data.copy()
location_world = location_world.drop(columns=['variant', 'num_sequences', 'date'])
location_world = location_world.drop_duplicates(subset = ['location'])
location_world.reset_index(drop=True, inplace=True)
```

```python
def ols_world(data):
    world_data_var = data.copy()

    for i in range(len(week)):
        world_data_var.replace(to_replace = week['date'][i], value = (i + 1), inplace= True)
        
    result = sm.ols(formula="num_sequences ~ location + date", data=world_data_var).fit()
    print(result.summary())
    
    for i in range(len(location_world['location'])):
        world_data_var.replace(to_replace = location_world['location'][i], value = (i + 1), inplace= True)
    sns.heatmap(world_data_var.corr(), vmin=-1, vmax=1, annot=True, cmap="Blues_r", cbar=False).set_title("Correlation Analysis", fontdict={'fontsize':12}, pad=12);
    
```

```python
# world variant of concern cases OLS Regression Result and correlation analysis
ols_world(world_data)
```

```python
# World ALPHA variant OLS Regression Result and correlation analysis
ols_world(world_data_alpha)
```

```python
# World BETA variant OLS Regression Result and correlation analysis
ols_world(world_data_beta)
```

```python
# World GAMMA variant OLS Regression Result and correlation analysis
ols_world(world_data_gamma)
```

```python
# World DELTA variant OLS Regression Result and correlation analysis
ols_world(world_data_delta)
```

#### OLS Regression Analysis
From this analysis observed that most of the p>0.05 and tvalue close to 0. This might be due to the fact that some country do not have sufficient sample to do the analysis. If we observe each variant regression result, we can be sure that spreading on some location has a similar transmission pattern like the Europe for all COVID-19 Variant of Concern and the date of time also matter since all of them have p < 0.05 and -1 < t or t > 1. 

If we observe closer, some countries give more samples on certain variants, thus strengthening the fact that during an outbreak, the location is related to the number of cases.

#### Correlation Analysis
Based on this analysis it is further strengthen the fact that the time frame matter and shown on the world_data_delta, gamma, and world_data heatmap. Therefore, we need to make further correlation analysis and separate the data based on a certain time frame.

```python
def corr_world(data):
    world_data_var = data.copy()

    for i in range(len(week)):
        world_data_var.replace(to_replace = week['date'][i], value = (i + 1), inplace= True)
    
    for i in range(len(location_world['location'])):
        world_data_var.replace(to_replace = location_world['location'][i], value = (i + 1), inplace= True)
        
    fig, axs = plt.subplots(nrows = 1, ncols = 3, figsize = (25,6))
    sns.heatmap(world_data_var.loc[(world_data_var['date'] >= 1) & (world_data_var['date'] <= 13)].corr(), vmin=-1, vmax=1, annot=True, ax = axs[0], cmap="Blues_r", cbar=False).set_title("Week 1 to 13", fontdict={'fontsize':20}, pad=12);
    sns.heatmap(world_data_var.loc[(world_data_var['date'] >= 14) & (world_data_var['date'] <= 27)].corr(), vmin=-1, vmax=1, annot=True, ax = axs[1], cmap="Blues_r", cbar=False).set_title("Week 14 to 27", fontdict={'fontsize':20}, pad=12);
    sns.heatmap(world_data_var.loc[(world_data_var['date'] >= 28) & (world_data_var['date'] <= 39)].corr(), vmin=-1, vmax=1, annot=True, ax = axs[2], cmap="Blues_r", cbar=False).set_title("Week 28 to 39", fontdict={'fontsize':20}, pad=12);
```

```python
# World variant of concern correlation analysis during different time frame
corr_world(world_data)
```
